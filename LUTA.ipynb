```python
#simulação de luta entre kyo e ken
#ler numeros inteiros

def ler_numero(mensagem):
    while True:
        try:
            return int(input(mensagem))
        except ValueError:
            print("Erro: digite apenas números inteiros.")

#vida dos personagens
hp_kyo = int(input('digite o hp do kyo:'))

hp_ken = int(input('digite o hp do ken:'))

#quantidade de golpes
golpes_kyo = 0

golpes_ken = 0

#valor do golpe


#porradaria
while hp_kyo > 0 and hp_ken > 0:
    golpe = int(input('digite o valor do golpe:'))

    if golpe > 0:
        hp_ken = hp_ken - golpe
        print('ryo agrediu o ken com', golpe, 'de dano!!!')
        print('vida atual do kyo:', hp_kyo)
        print('vida atual do ken:', hp_ken)

        golpes_kyo += 1

    elif golpe < 0 :
        hp_kyo = hp_kyo - abs(golpe)
        print('ken agrediu o kyo com', abs(golpe), 'de dano!!!')
        print('vida atual do kyo:', hp_kyo)
        print('vida atual do ken:', hp_ken)
        golpes_ken += 1

if hp_kyo <= 0:
    print('ken venceu a luta!!!')
    print('golpes desferidos por kyo:', golpes_kyo)
    print('golpes desferidos por ken:', golpes_ken)

elif hp_ken <= 0:
    print('kyo venceu a luta!!!')
    print('golpes desferidos por kyo:', golpes_kyo)
    print('golpes desferidos por ken:', golpes_ken)
```

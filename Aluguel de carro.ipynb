```python
#calculo de aluguel de carro

diaria = 60

km_rodado = 0.15


print('bem vindo ao seu relatorio de aluguel do seu veiculo!!!')

dias = int(input('digite a quantidade de dias que voce ficou com o veiculo: '))

km = float(input('digite quantos km voce percorreu: '))

diaria_final = dias * diaria

km_final = km*km_rodado

total= diaria_final + km_final

print('voce rodou {} dias com o carro e percorreu {:.2f}KM.'.format(dias,km))

print('o valor a se pagar é de:R$',total )
```

```python
import random
```

```python
#TERMO

nomes = ['Achar', 'Andar', 'Bater','Beber','Caber','Catar','Criar','Dizer'
        ,'Dotar','Estar','Fazer','Falar','Gelar','Gerar','Levar','Ouvir','Parar','Poder','Saber','Viver']


nomes_aleatorio = random.choice(nomes).upper()


tentativas = 6

print(nomes_aleatorio)
print('-'*60)
print('Descubra a palavra com 5 letras! voce tem 6 tentaivas!!')

for i in range(1,7):
    while True:
        chute = input("\nTentativa {}/{}: ".format(i,tentativas)).upper()

        if len(chute) == 5 and chute.isalpha():
            break

        print('palavra invalida, digite apenas palavras com 5 letras sem numeros ou caracteres')


    validação = ""

    for posição, letra in enumerate(chute):


        if letra == nomes_aleatorio[posição]:

            validação = validação + '\033[42;30m {} \033[0m'.format(letra)
        
        elif letra in nomes_aleatorio:

            validação = validação + '\033[43;30m {} \033[0m'.format(letra) 

        else:

            validação = validação + '\033[47;30m {} \033[0m'.format(letra)

    print(validação)

    if chute == nomes_aleatorio:
        print("voce acertou!!!!!!")

else:
    print('a palavra correta era : {}'.format(nomes_aleatorio))
        



```

```python
#calculo de tinta

altura = float(input('qual a altura da parede?: '))

largura = float(input('qual a largura da parede?: '))

area = altura * largura

tinta = area / 2

print('sua parede tem a dimensão de {}x{} e sua areea é de {:.2f}'.format(altura, largura, area))

print('voce precisará de {:.2f}L de tinta'.format(tinta))
```

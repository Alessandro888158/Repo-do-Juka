```python
limite = 80

multa = 7.00



velocidade = int(input('digite a velocidade do carro: '))

if velocidade > limite:
    print('voce ultrapassou o limite de velocidade!')
    valor = (velocidade-limite)*multa
    print('voce foi multado em {} Reais'.format(valor))

else:
    print('voce esta dentro do limite de velocidade, diriga com cuidado')

```

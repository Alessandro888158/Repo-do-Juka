```python
#Calculo de Sálario Líquido

salario_Bruto = float(input("Digite o salário bruto: R$ "))

#Descontos INSS

if  salario_Bruto <= 1045.00:
    inss = salario_Bruto * 7.5 / 100

elif salario_Bruto >1045.01 and salario_Bruto <= 2089.60:
    inss = salario_Bruto * 9 / 100

elif salario_Bruto >2089.61 and salario_Bruto <= 3134.40:
    inss = salario_Bruto * 12 / 100

elif salario_Bruto >3134.41:   
    inss = salario_Bruto * 14 / 100

#calculo do Imposto de renda

if salario_Bruto <= 1903.98:
    imposto_renda = 0 

elif salario_Bruto > 1903.99 and salario_Bruto <= 2826.65:
    imposto_renda = (salario_Bruto * 7.5 / 100) - 142.80

elif salario_Bruto > 2826.66 and 3751.05:
    imposto_renda = (salario_Bruto *15 / 100) - 354.80

elif salario_Bruto > 3751.06 and salario_Bruto <= 4664.68:
    imposto_renda = (salario_Bruto * 22.5 / 100) - 636.13

elif salario_Bruto > 4664.68:
    imposto_renda = (salario_Bruto * 27.5 / 100) - 869.36

salario_Liquido = salario_Bruto - inss - imposto_renda

print("o salario bruto é: ", salario_Bruto)
print("o desconto do INSS é: ", inss)
print("o desconto do imposto de renda é: ", imposto_renda)
print("o valor total do salario luquido é: ", salario_Liquido)


    
```
